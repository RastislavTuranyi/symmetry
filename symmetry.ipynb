```python
def reduction(reducible, group):
  result = group.iloc[1:] * group.iloc[0] * np.array(reducible)

  result['number of appearences'] = result.sum(1) / group.iloc[0].sum()
  return result.astype(int)


def Elevel(group, *args):
  reducible = group.loc[[*args]].product().rename(' × '.join(args))  #calculate the product of the inputted rows

  if (group == reducible).all(1).any() == True:  #if the product is an irreducible representation
    temp = (group == reducible).all(1)  #pd series indicating which irreducible representation the product is
    print(temp[temp == True].index.values[0])
    return reducible
  
  else:
    result = reduction(reducible, group)
    return result.append(reducible)

Elevel(D2h, 'B2u', 'B1u')
```
